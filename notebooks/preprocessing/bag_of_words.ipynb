```python
# import required packages
import pandas as pd
import numpy as np

# preprocess
from sklearn.feature_extraction.text import CountVectorizer

# local scripts
from text_utils import preprocess_corpus
```

```python
df = pd.read_csv('../../data/train_data.csv')

# dataframe shape: (rows, columns)
display(df.shape)

# first 5 datapoints
display(df.head())
```

```python
# preprocess documents
# remove special characters, stopwords
# lemmatization
clean_sentence = preprocess_corpus(df.Sentence)

clean_sentence.head()
```

```python
# instantiate a CountVectorizer
cv = CountVectorizer(min_df=4, max_df=0.01)

# train and construct bag of words
counts = cv.fit_transform(clean_sentence)
```

```python
# extract bag of words as dataframe
df_bow = pd.DataFrame(counts.A, columns=cv.get_feature_names_out())

# include Sentiment column
df_bow['Sentiment'] = df['Sentiment']

# shape
display(df_bow.shape)

# first 5 datapoints
df_bow.head()
```
